```
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE

# Load your cleaned data
df = pd.read_csv("solar_flare_data_cleaned.csv")

df = df[df['flare_class_peak'] != 'A']

# Drop any missing values (important before SMOTE)
df = df.dropna()

# Encode class labels
le = LabelEncoder()
df['class_encoded'] = le.fit_transform(df['flare_class_peak'])

# Separate features and labels
X = df[['xrsb_flux', 'background_flux', 'integrated_flux']]
y = df['class_encoded']

from imblearn.over_sampling import SMOTE

# Create the SMOTE object
smote = SMOTE(random_state=42)

# Resample the data
X_resampled, y_resampled = smote.fit_resample(X, y)

# Inverse transform to get back class labels
y_resampled_labels = le.inverse_transform(y_resampled)

# Combine features and labels into a DataFrame
df_resampled = pd.DataFrame(X_resampled, columns=X.columns)
df_resampled['flare_class_peak'] = y_resampled_labels

df_resampled.to_csv("solar_flare_data_smote.csv", index=False)
print(df_resampled['flare_class_peak'].value_counts())
```
