```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load cleaned data
df = pd.read_csv("solar_flare_data_smote.csv")

# Count the number of flares in each class
class_counts = df['flare_class_peak'].value_counts().sort_index()

# Plot bar chart
plt.figure(figsize=(6, 4))
sns.barplot(x=class_counts.index, y=class_counts.values)

# Add labels and title
plt.title('Number of Solar Flares by Class')
plt.xlabel('Flare Class (B, C, M, X)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()


# Load cleaned data
df = pd.read_csv("solar_flare_data_smote.csv")

# Set desired class order
flare_class_order = ['A', 'B', 'C', 'M', 'X']


# Xrsb_flux vs. flare_class
plt.figure(figsize=(8, 5))
sns.boxplot(x='flare_class_peak', y='xrsb_flux', data=df, order=flare_class_order)

# Customize plot
plt.title('Xrsb Flux by Flare Class at Peak')
plt.xlabel('Flare Class (Peak)')
plt.ylabel('Xrsb Flux')
plt.yscale('log')  # Apply log scale to better visualize wide flux ranges

# Show plot
plt.tight_layout()
plt.show()

# Background_flux vs. flare_class

plt.figure(figsize=(8, 5))
sns.boxplot(x='flare_class_peak', y='background_flux', data=df, order=flare_class_order)

# Customize plot
plt.title('Background Flux by Flare Class at Peak')
plt.xlabel('Flare Class (Peak)')
plt.ylabel('Background Flux')
plt.yscale('log')

# Show plot
plt.tight_layout()
plt.show()

# Integrated_flux vs. flare_class

plt.figure(figsize=(8, 5))
sns.boxplot(x='flare_class_peak', y='integrated_flux', data=df, order=flare_class_order)

# Customize plot
plt.title('Integrated Flux by Flare Class at Peak')
plt.xlabel('Flare Class (Peak)')
plt.ylabel('Integrated Flux')
plt.yscale('log')

# Show plot
plt.tight_layout()
plt.show()
```
